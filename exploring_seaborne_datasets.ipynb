```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips_dataset = sns.load_dataset('tips')
```

```python
tips_dataset.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_dataset)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_dataset, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips_dataset)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_dataset)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips_dataset, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_dataset)#, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_dataset, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips_dataset, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_dataset, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_dataset, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips_dataset, jitter=True, hue='sex', split=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips_dataset)
```

```python
sns.violinplot(x= 'day', y= 'total_bill', data= tips_dataset)
sns.swarmplot(x= 'day', y= 'total_bill', data= tips_dataset, color= 'black')
```

```python
sns.factorplot(x= 'day', y= 'total_bill', data= tips_dataset, kind='bar')
```

```python
sns.FacetGrid(data= tips_dataset, row= 'time', col= 'smoker').map(plt.hist, 'total_bill')
```

```python
graph= sns.FacetGrid(data= tips_dataset, row= 'time', col= 'smoker')
graph.map(plt.scatter, 'total_bill', 'tip', edgecolor= 'w')
# Here we explore smokers vs non-smokers paying tips during lunch vs dinner
```
